```python
import matplotlib.pyplot as plt
import networkx as nx

nodes_list = [0,1,2,3,4,5,6]
edges_list = [(1,0),(2,3),(1,3),(3,5),(4,5),(5,6)]
is_strongly_connected = False

# Inititalize graph
G = nx.Graph()
# Get petersen graph from library
#G = nx.complete_graph(4)
# Custom graph
G.add_nodes_from(nodes_list)
G.add_edges_from(edges_list)

# Mark strongly connected nodes
node_color_map = []
for node in G:
    if len(list(nx.neighbors(G, node))) > 2:
        node_color_map.append('red')
        is_strongly_connected = True
    else:
        node_color_map.append('green')

# Print is this Graph strongly connected
if is_strongly_connected:
    print ('Graph is strongly connected')
else:
    print ('Graph is not strongly connected')

# Draw graph
nx.draw(G, node_color=node_color_map, with_labels=True, font_weight='bold')
plt.show()
```
